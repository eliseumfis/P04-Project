```python
# ============================================================================
# ANÁLISIS COMPLETO DE ESTRELLA VARIABLE - CON EJECUCIÓN
# ============================================================================

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import warnings
from matplotlib.gridspec import GridSpec
from datetime import datetime

warnings.filterwarnings('ignore')

# Configuración de matplotlib
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 12

# ============================================================================
# 1. CARGAR DATOS DESDE TUS ARCHIVOS
# ============================================================================

print("="*80)
print("CARGANDO DATOS DESDE ARCHIVOS...")
print("="*80)

# Cargar datos principales
main_data = np.loadtxt("Data.dat")

time = main_data[:, 0]
observed = main_data[:, 1]
calculated = main_data[:, 2]
residuals = main_data[:, 3]
point_weight = main_data[:, 4]
point_error = main_data[:, 5]

print(f"Datos principales cargados: {len(time)} puntos")
print(f"Tiempo: {time[0]:.4f} a {time[-1]:.4f} ({time[-1]-time[0]:.2f} unidades)")
print(f"Flujo observado: media={np.mean(observed):.6f}, std={np.std(observed):.6f}")

# Cargar periodos
periodos = np.loadtxt("freqs.per")
print(f"\nFrecuencias cargadas: {len(periodos)} períodos encontrados")

# Organizar frecuencias
frecuencias = {}
for i in range(min(6, len(periodos))):
    frecuencias[f'f{i+1}'] = {
        'freq': periodos[i, 0],
        'amp': periodos[i, 1],
        'phase': periodos[i, 2]
    }
    print(f"  f{i+1}: {periodos[i, 0]:.6f} c/d, amp={periodos[i, 1]:.6f}")

# Cargar transformadas de Fourier
print("\nCargando transformadas de Fourier...")
fourier_data = {}

# Cargar cada archivo con manejo de errores
files_to_load = [
    ("TF_1.fou", "original"),
    ("TF_2.fou", "residual_1"),
    ("TF_3.fou", "residual_2"),
    ("TF_res_1.fou", "residual_3"),
    ("TF_res_2.fou", "residual_4"),
    ("TF_res_3.fou", "residual_5"),
    ("Spectral_window.fou", "spectral_window"),
    ("f1_ffwsw.fou", "f1_folded"),
    ("f2_ffwsw.fou", "f2_folded"),
    ("f3_ffwsw.fou", "f3_folded"),
    ("f4_ffwsw.fou", "f4_folded"),
    ("f5_ffwsw.fou", "f5_folded"),
    ("f6_ffwsw.fou", "f6_folded"),
    ("Spectrum_noise_original_data.dat", "noise_spectrum"),
    ("Spectrum_noise_residuals.dat", "noise_spectrum_residuals")
]

for filename, key in files_to_load:
    try:
        data = np.loadtxt(filename)
        fourier_data[key] = {'freq': data[:, 0], 'amp': data[:, 1]}
        print(f"  ✓ {filename} cargado como '{key}'")
    except Exception as e:
        print(f"  ✗ No se pudo cargar {filename}: {e}")

# Cargar ruido en frecuencias
try:
    freq_noises = np.loadtxt("Noise_frequencies.dat")
    for i, key in enumerate(list(frecuencias.keys())[:min(6, len(freq_noises))]):
        if i < len(freq_noises):
            frecuencias[key]['noise_amp'] = freq_noises[i, 1]
            frecuencias[key]['snr'] = freq_noises[i, 2]
    print("✓ Noise_frequencies.dat cargado")
except:
    print("✗ No se pudo cargar Noise_frequencies.dat")

# ============================================================================
# 2. ANÁLISIS ESTADÍSTICO
# ============================================================================

print("\n" + "="*80)
print("ANÁLISIS ESTADÍSTICO")
print("="*80)

# Estadísticas básicas
tiempo_total = time[-1] - time[0]
dt_mean = np.mean(np.diff(time))

obs_mean = np.mean(observed)
obs_std = np.std(observed)
obs_range = np.max(observed) - np.min(observed)

res_mean = np.mean(residuals)
res_std = np.std(residuals)
res_skew = stats.skew(residuals)
res_kurtosis = stats.kurtosis(residuals)

var_obs = np.var(observed)
var_res = np.var(residuals)
var_reduction = (var_obs - var_res) / var_obs * 100 if var_obs > 0 else 0

# Test de normalidad
stat, p_value = stats.normaltest(residuals)

print(f"\n1. ESTADÍSTICAS TEMPORALES:")
print(f"   Duración total: {tiempo_total:.4f} unidades")
print(f"   Intervalo promedio: {dt_mean:.6f}")

print(f"\n2. ESTADÍSTICAS DE FLUJO:")
print(f"   Media: {obs_mean:.6f}")
print(f"   Desviación estándar: {obs_std:.6f}")
print(f"   Rango: {obs_range:.6f}")

print(f"\n3. ESTADÍSTICAS DE RESIDUOS:")
print(f"   Media: {res_mean:.6f}")
print(f"   Desviación estándar: {res_std:.6f}")
print(f"   Asimetría: {res_skew:.3f}")
print(f"   Curtosis: {res_kurtosis:.3f}")

print(f"\n4. CALIDAD DEL AJUSTE:")
print(f"   Varianza observada: {var_obs:.6e}")
print(f"   Varianza residuos: {var_res:.6e}")
print(f"   Reducción de varianza: {var_reduction:.2f}%")

print(f"\n5. TEST DE NORMALIDAD:")
print(f"   p-value: {p_value:.6f}")
print(f"   ¿Distribución normal? {'Sí' if p_value > 0.05 else 'No'}")

# Guardar estadísticas
estadisticas = {
    'tiempo_total': tiempo_total,
    'obs_stats': {'mean': obs_mean, 'std': obs_std, 'range': obs_range},
    'res_stats': {'mean': res_mean, 'std': res_std, 'skewness': res_skew, 'kurtosis': res_kurtosis},
    'var_reduction': var_reduction,
    'normalidad_pvalue': p_value
}

# ============================================================================
# 3. ANÁLISIS DE PERIODICIDADES Y SNR
# ============================================================================

print("\n" + "="*80)
print("ANÁLISIS DE PERIODICIDADES Y SNR")
print("="*80)

resultados = []

for key, freq_data in frecuencias.items():
    freq = freq_data['freq']
    amp = freq_data['amp']
    snr = freq_data.get('snr', None)
    
    periodo = 1/freq if freq != 0 else float('inf')
    
    # Calcular FAP
    if snr is not None:
        N_ind = len(time) / 2
        z = snr**2 / 2
        fap_exacta = 1 - (1 - np.exp(-z))**N_ind
        fap_approx = N_ind * np.exp(-z)
        fap = min(fap_exacta, fap_approx, 1.0)
        prob_real = (1 - fap) * 100
        
        if snr >= 4.0:
            significancia = "MUY SIGNIFICATIVA"
        elif snr >= 3.0:
            significancia = "MODERADAMENTE SIGNIFICATIVA"
        else:
            significancia = "NO SIGNIFICATIVA"
    else:
        snr = "N/A"
        fap = "N/A"
        prob_real = "N/A"
        significancia = "N/A"
    
    resultado = {
        'frecuencia': key,
        'freq_value': freq,
        'periodo': periodo,
        'amplitud': amp,
        'snr': snr,
        'fap': fap,
        'prob_real': prob_real,
        'significancia': significancia
    }
    
    resultados.append(resultado)
    
    print(f"\n{key.upper()}:")
    print(f"  Frecuencia: {freq:.6f} c/d")
    print(f"  Período: {periodo:.6f} unidades")
    print(f"  Amplitud: {amp:.6f}")
    if snr != "N/A":
        print(f"  SNR: {snr:.3f}")
        print(f"  FAP: {fap:.6f}")
        print(f"  Prob. real: {prob_real:.2f}%")
        print(f"  Significancia: {significancia}")

# Contar significativas
significativas = [r for r in resultados if isinstance(r['snr'], (int, float)) and r['snr'] >= 4.0]
print(f"\n✓ Frecuencias significativas (SNR ≥ 4.0): {len(significativas)} de {len(resultados)}")

# ============================================================================
# 4. GENERACIÓN DE GRÁFICOS
# ============================================================================

print("\n" + "="*80)
print("GENERANDO GRÁFICOS")
print("="*80)

# GRÁFICO 1: CURVA DE LUZ COMPLETA
print("Generando Gráfico 1: Curva de luz completa...")
fig1, axes1 = plt.subplots(2, 2, figsize=(14, 10))

# Subplot 1: Curva de luz
axes1[0, 0].scatter(time, observed, s=5, alpha=0.6, color='blue', label='Observado')
axes1[0, 0].plot(time, calculated, 'r-', linewidth=1, alpha=0.8, label='Modelo')
axes1[0, 0].set_xlabel('Tiempo')
axes1[0, 0].set_ylabel('Flujo/Magnitud')
axes1[0, 0].set_title('Curva de Luz Completa')
axes1[0, 0].legend()
axes1[0, 0].grid(True, alpha=0.3)

# Subplot 2: Residuos
axes1[0, 1].scatter(time, residuals, s=5, alpha=0.6, color='green')
axes1[0, 1].axhline(y=0, color='r', linestyle='--', linewidth=1)
axes1[0, 1].set_xlabel('Tiempo')
axes1[0, 1].set_ylabel('Residuos')
axes1[0, 1].set_title('Residuos del Ajuste')
axes1[0, 1].grid(True, alpha=0.3)

# Subplot 3: Histograma de residuos
axes1[1, 0].hist(residuals, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')
mu, sigma = stats.norm.fit(residuals)
x = np.linspace(residuals.min(), residuals.max(), 100)
pdf = stats.norm.pdf(x, mu, sigma)
axes1[1, 0].plot(x, pdf, 'r-', linewidth=2, label=f'Normal: μ={mu:.3f}, σ={sigma:.3f}')
axes1[1, 0].set_xlabel('Residuos')
axes1[1, 0].set_ylabel('Densidad')
axes1[1, 0].set_title('Distribución de Residuos')
axes1[1, 0].legend()
axes1[1, 0].grid(True, alpha=0.3)

# Subplot 4: Periodograma
if 'original' in fourier_data:
    axes1[1, 1].plot(fourier_data['original']['freq'], fourier_data['original']['amp'], 
                     'b-', linewidth=0.8, alpha=0.7)
    
    # Marcar frecuencias
    colors = ['red', 'orange', 'green', 'purple', 'brown', 'pink']
    for i, (key, freq_data) in enumerate(list(frecuencias.items())[:6]):
        freq_val = freq_data['freq']
        axes1[1, 1].axvline(x=freq_val, color=colors[i%6], linestyle='--', 
                           linewidth=1, alpha=0.7, label=f'{key}: {freq_val:.3f}')
    
    axes1[1, 1].set_xlabel('Frecuencia (c/d)')
    axes1[1, 1].set_ylabel('Amplitud')
    axes1[1, 1].set_title('Periodograma Original')
    axes1[1, 1].legend(fontsize=8)
    axes1[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('figura_1_analisis_completo.png', dpi=150, bbox_inches='tight')
print("✓ Figura 1 guardada: 'figura_1_analisis_completo.png'")
plt.close(fig1)

# GRÁFICO 2: CURVAS PLEGADAS
print("Generando Gráfico 2: Curvas plegadas...")
n_frecuencias_plot = min(4, len(frecuencias))
fig2, axes2 = plt.subplots(2, 2, figsize=(12, 8))
axes2 = axes2.flatten()

for i, (key, freq_data) in enumerate(list(frecuencias.items())[:n_frecuencias_plot]):
    freq = freq_data['freq']
    periodo = 1/freq
    
    # Plegar curva
    folded_time = (time / periodo) % 1
    sorted_idx = np.argsort(folded_time)
    
    axes2[i].scatter(folded_time[sorted_idx], observed[sorted_idx], 
                    s=8, alpha=0.6, color=plt.cm.tab10(i))
    axes2[i].set_xlabel('Fase')
    axes2[i].set_ylabel('Flujo/Magnitud')
    axes2[i].set_title(f'{key}: P = {periodo:.6f}')
    axes2[i].grid(True, alpha=0.3)

# Ocultar ejes extras si hay menos de 4 frecuencias
for i in range(n_frecuencias_plot, 4):
    axes2[i].axis('off')

plt.tight_layout()
plt.savefig('figura_2_curvas_plegadas.png', dpi=150, bbox_inches='tight')
print("✓ Figura 2 guardada: 'figura_2_curvas_plegadas.png'")
plt.close(fig2)

# GRÁFICO 3: COMPARACIÓN DE SNR
print("Generando Gráfico 3: Comparación de SNR...")
resultados_con_snr = [r for r in resultados if isinstance(r['snr'], (int, float))]

if len(resultados_con_snr) > 0:
    fig3, ax3 = plt.subplots(figsize=(10, 6))
    
    nombres = [r['frecuencia'] for r in resultados_con_snr]
    snr_values = [r['snr'] for r in resultados_con_snr]
    
    bars = ax3.bar(nombres, snr_values, color=plt.cm.RdYlGn(np.array(snr_values)/max(snr_values)))
    ax3.axhline(y=4.0, color='red', linestyle='--', linewidth=2, label='Umbral SNR=4.0')
    ax3.axhline(y=3.0, color='orange', linestyle=':', linewidth=2, label='Umbral SNR=3.0')
    
    # Añadir valores
    for bar, snr in zip(bars, snr_values):
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{snr:.2f}', ha='center', va='bottom')
    
    ax3.set_xlabel('Frecuencia')
    ax3.set_ylabel('SNR')
    ax3.set_title('Relación Señal-Ruido por Frecuencia')
    ax3.legend()
    ax3.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.savefig('figura_3_snr_comparacion.png', dpi=150, bbox_inches='tight')
    print("✓ Figura 3 guardada: 'figura_3_snr_comparacion.png'")
    plt.close(fig3)

# GRÁFICO 4: VENTANA ESPECTRAL
print("Generando Gráfico 4: Ventana espectral...")
if 'spectral_window' in fourier_data:
    fig4, ax4 = plt.subplots(figsize=(10, 6))
    
    ax4.plot(fourier_data['spectral_window']['freq'], 
            fourier_data['spectral_window']['amp'], 
            'b-', linewidth=1, alpha=0.8)
    ax4.set_xlabel('Frecuencia (c/d)')
    ax4.set_ylabel('Amplitud')
    ax4.set_title('Ventana Espectral')
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('figura_4_ventana_espectral.png', dpi=150, bbox_inches='tight')
    print("✓ Figura 4 guardada: 'figura_4_ventana_espectral.png'")
    plt.close(fig4)

# Lista de figuras generadas
figuras_generadas = [
    'figura_1_analisis_completo.png',
    'figura_2_curvas_plegadas.png',
    'figura_3_snr_comparacion.png',
    'figura_4_ventana_espectral.png'
]

# ============================================================================
# 5. IDENTIFICACIÓN DEL TIPO DE ESTRELLA
# ============================================================================

print("\n" + "="*80)
print("IDENTIFICACIÓN DEL TIPO DE ESTRELLA")
print("="*80)

# Obtener la mejor frecuencia
if len(resultados_con_snr) > 0:
    resultados_con_snr.sort(key=lambda x: x['snr'], reverse=True)
    mejor = resultados_con_snr[0]
    
    periodo_principal = mejor['periodo']
    amplitud_principal = mejor['amplitud']
    snr_principal = mejor['snr']
    
    print(f"\nPARÁMETROS PRINCIPALES:")
    print(f"  Período: {periodo_principal:.6f} unidades")
    print(f"  Amplitud: {amplitud_principal:.6f}")
    print(f"  SNR: {snr_principal:.2f}")
    
    # Determinar tipo basado en período y amplitud
    if periodo_principal < 0.3:
        if amplitud_principal < 0.3:
            tipo = "Delta Scuti o SX Phoenicis"
            probabilidad = 70
        else:
            tipo = "RR Lyrae de período corto"
            probabilidad = 60
    elif periodo_principal < 1.0:
        if amplitud_principal > 0.5:
            tipo = "RR Lyrae"
            probabilidad = 80
        else:
            tipo = "Variable de largo período tipo Mira"
            probabilidad = 50
    elif periodo_principal < 10:
        tipo = "Cefeida clásica"
        probabilidad = 75
    else:
        tipo = "Variable de largo período"
        probabilidad = 65
    
    # Ajustar por SNR
    if snr_principal < 4.0:
        tipo = "Posible ruido o variabilidad débil"
        probabilidad = 40
    elif snr_principal < 3.0:
        tipo = "Probablemente ruido"
        probabilidad = 20
    
    print(f"\nCONCLUSIÓN:")
    print(f"  Tipo más probable: {tipo}")
    print(f"  Probabilidad estimada: {probabilidad}%")
    
    tipo_estrella = {
        'tipo': tipo,
        'probabilidad': probabilidad,
        'periodo_principal': periodo_principal,
        'amplitud_principal': amplitud_principal,
        'snr_principal': snr_principal
    }
else:
    tipo_estrella = {
        'tipo': "Indeterminado (sin SNR)",
        'probabilidad': 0,
        'periodo_principal': 0,
        'amplitud_principal': 0,
        'snr_principal': 0
    }
```
